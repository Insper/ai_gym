```python
from aigyminsper.search.SearchAlgorithms import BuscaLargura
from aigyminsper.search.SearchAlgorithms import BuscaProfundidade
from aigyminsper.search.SearchAlgorithms import BuscaProfundidadeIterativa
from aigyminsper.search.Graph import State
from datetime import datetime
import pandas as pd
```

```python
class SumOne(State):

    def __init__(self, n, op, g):
        self.operator = op
        self.number = n
        self.goal = g

    def successors(self):
        successors = []
        if self.number < self.goal:
            successors.append(SumOne(self.number+1, "+1 ", self.goal))
            successors.append(SumOne(self.number+2, "+2 ", self.goal))
        return successors

    def is_goal(self):
        if self.goal == self.number:
            return True
        return False

    def description(self):
        return "Este Ã© um agente simples que sabe somar 1 e 2"

    def cost(self):
        return 1

    def env(self):
        return self.number
```

```python
objetivo = 50
```

## Busca Profundidade Iterativa

```python
algorithm = BuscaProfundidadeIterativa()
df = pd.DataFrame(columns=['Algoritmo', 'Objetivo', 'Tempo de processamento'])

for i in range(0, objetivo+1):
    state = SumOne(1, '', i)
    start_time = datetime.now()
    result = algorithm.search(state)
    end_time = datetime.now()
    df.loc[len(df)] = [str(algorithm), objetivo, end_time - start_time]
    print('Algoritmo: ' + algorithm.__class__.__name__)
    print('Objetivo: ' + str(i))
    print('Tempo de processamento: ' + str(end_time - start_time))

df.to_csv('buscaprofundidadeiterativa.csv', index=False)
```

## Busca Largura

```python
state = SumOne(1, '', objetivo)
algorithms = [BuscaLargura()]
df = pd.DataFrame(columns=['Algoritmo', 'Objetivo', 'Tempo de processamento'])

for algorithm in algorithms:
    for i in range(0, objetivo + 1):
        state = SumOne(1, '', i)
        start_time = datetime.now()
        result = algorithm.search(state)
        end_time = datetime.now()
        df.loc[len(df)] = [str(algorithm), objetivo, end_time - start_time]
        print('Algoritmo: ' + algorithm.__class__.__name__)
        print('Objetivo: ' + str(i))
        print('Tempo de processamento: ' + str(end_time - start_time))
df.to_csv('buscalargura.csv', index=False)
```

```python
state = SumOne(1, '', objetivo)
algorithms = [BuscaProfundidade()]
ms = [10, 100]
df = pd.DataFrame(columns=['Algoritmo', 'Objetivo', 'Tempo de processamento'])

for algorithm in algorithms:
    for x in ms:
        for i in range(0, objetivo+1):
            state = SumOne(1, '', i)
            start_time = datetime.now()
            result = algorithm.search(state, m=x)
            end_time = datetime.now()
            df.loc[len(df)] = [str(algorithm) + ' m=' + str(x), objetivo, end_time - start_time]
            print('Algoritmo: ' + algorithm.__class__.__name__ + ' m=' + str(x))
            print('Objetivo: ' + str(i))
            print('Tempo de processamento: ' + str(end_time - start_time))

df.to_csv('buscaprofundidade.csv', index=False)
```
